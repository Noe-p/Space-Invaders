```python
try:  # import as appropriate for 2.x vs. 3.x
    import tkinter as tk
    import tkinter.messagebox as tkMessageBox
except:
    import Tkinter as tk
    import tkMessageBox


class Example:
    def __init__(self):
        self.root = tk.Tk()
        self.rect_id = None
        self.canvas_width = 600
        self.canvas_height = 400
        self.square_width = 50
        self.canvas = tk.Canvas(self.root, width=self.canvas_width, height=self.canvas_height)
        self.canvas.pack()
        
    def install(self):
        w, x, y = 50, 0, 0
        self.rect_id = self.canvas.create_rectangle(self.canvas_width/2,self.canvas_height/2, (self.canvas_width/2) + w ,(self.canvas_height/2)+ w,outline='black',fill='red')
        
    def keypress(self, event):
        # A MODIFIER
        d=10
        if event.keysym == 'Left':
            self.canvas.move(self.rect_id, -d, 0) 
           
            
        elif event.keysym == 'Right':
            self.canvas.move(self.rect_id, d, 0) 

        elif event.keysym == 'space':
            self.canvas.move(self.rect_id, 0, -d) 
       
            
    def start(self):
        self.install()
        # Quand n'importe quelle touche du clavier est appuyée, exécuter keypress
        self.root.bind("<Key>", self.keypress)
        self.root.mainloop()

ex = Example()
ex.start()
```

```python
try:  # import as appropriate for 2.x vs. 3.x
    import tkinter as tk
    import tkinter.messagebox as tkMessageBox
except:
    import Tkinter as tk
    import tkMessageBox


class Example:
    def __init__(self):
        self.root = tk.Tk()
        self.rect_id = None
        self.canvas_width = 600
        self.canvas_height = 400
        self.square_width = 50
        self.gap = 20
        self.canvas = tk.Canvas(self.root, width=self.canvas_width, height=self.canvas_height)
        self.canvas.pack()

    def install(self):
        # creation du rectangle
        w, x, y = 50, 0, 0

        self.rect_id = self.canvas.create_rectangle(x,y, x + w ,y + w,outline='black',fill='red')


    def start_animation(self):
        # execute self.animation dans 10ms
        self.canvas.after(10, self.animation)

    def animation(self):
        max_w = self.canvas_width
        # calcul du deplacement dans self.gap en tenant compte de self.canvas_width
        self.gap = 20
        x1, y1, x2, y2 = self.canvas.bbox(self.rect_id)

        print(max_w, x1, x2)
        # deplacement du rectangle
        if x2 > max_w:
            self.canvas.move(self.rect_id, -self.canvas_width, 0) 

        self.canvas.move(self.rect_id, self.gap, 0)
        # execute a nouveau self.animation dans 300ms
        self.canvas.after(300, self.animation)

    def start(self):
        self.install()
        self.start_animation()
        self.root.mainloop()


ex = Example()
ex.start()
```
